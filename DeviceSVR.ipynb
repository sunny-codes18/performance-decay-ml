```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv('laptop_performance_dataset.csv')
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
X=sc_x.fit_transform(X)
y=sc_y.fit_transform(y.reshape(len(y),1))
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='rbf')
regressor.fit(X_train,y_train.ravel())
```

```python
def predict_performance():
    print("\nEnter laptop parameters to predict performance:")

    # Ask for input for each feature
    hours_used = float(input("Hours Used: "))
    temperature = float(input("Temperature (°C): "))
    load = float(input("Load (0 to 1): "))
    voltage_fluctuation = float(input("Voltage Fluctuation (V): "))
    humidity = float(input("Humidity (%): "))
    maintenance_interval = float(input("Maintenance Interval (hours): "))
    fan_speed = float(input("Fan Speed (RPM): "))

    # Prepare input array
    user_input = np.array([[hours_used, temperature, load, voltage_fluctuation,
                            humidity, maintenance_interval, fan_speed]])

    # Scale input
    user_input_scaled = sc_x.transform(user_input)

    # Predict and inverse scale
    predicted_scaled = regressor.predict(user_input_scaled)
    predicted = sc_y.inverse_transform(predicted_scaled.reshape(-1, 1))

    print(f"\nPredicted Laptop Performance: {predicted[0][0]:.2f} / 100")

# Call the function
if __name__ == "__main__":
    predict_performance()
```
